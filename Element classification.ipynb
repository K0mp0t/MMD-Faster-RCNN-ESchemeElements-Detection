```python
import os

def get_relative_paths(root_path: str):
    root_content = os.listdir(root_path)
    relative_image_paths = []
    for element in root_content:
        if os.path.isdir(os.path.join(root_path, element)):
            relative_image_paths += get_relative_paths(os.path.join(root_path, element))
        else:
            relative_image_paths.append(os.path.join(root_path, element))
    return relative_image_paths
```

```python
images_path = f'E:\\Peter\\Scheme\\scheme_copy'

file_paths = get_relative_paths(images_path)
```

```python
for i in range(len(file_paths)):
   os.rename(file_paths[i], os.path.join(images_path, str(i)+'.png'))
```

```python
for path in os.listdir(images_path):
    current_path = os.path.join(images_path, path)
    if os.path.isdir(current_path) and len(os.listdir(current_path)) == 0:
        os.rmdir(current_path)
```
