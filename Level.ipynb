**Level transformation demo**

```python
%matplotlib inline

from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

im_path = 'E:/Peter/Scheme/scheme_copy/scheme_001.png'
image = Image.open(im_path)
```

```python
def level_grayscale_image(image, leveling_gap):
    assert image.mode == 'L'
    image_copy = np.array(image.copy())
    image_copy[image_copy < leveling_gap] = 0
    image_copy[image_copy > (255 - leveling_gap)] = 255
    
    return Image.fromarray(image_copy, 'L')
```

```python
image_copy = level_grayscale_image(image, 20)
fig, axes = plt.subplots(figsize = (20, 20))
axes.imshow(image_copy, cmap='gray')
```
